```python
import pandas as pd

df = pd.read_csv('2017_01_10440-01-01-2_臺中市高級中等學校教育經費概況(預、決算數).csv')

df.head()
```

```python
temp = df[['Complex2Name','FValue']]
temp
```

```python
Type = temp.groupby(['Complex2Name'])['FValue'].sum().reset_index()
```

```python
Type['year'] = 2017
Type
```

```python
df2018 = pd.read_csv('2018_01_10440-01-01-2_臺中市高級中等學校教育經費概況(預、決算數).csv')
```

```python
new = df2018[['Complex2Name','FValue']].groupby(['Complex2Name'])['FValue'].sum().reset_index()
```

```python
new['year'] = 2018
new
```

```python
frames = [Type, new]
result = pd.concat(frames)
result
```
